#### 4. Compute ngrams  

a. Find the 10 most common bigrams  
b. Find the 10 most common trigrams  

```python
# a)
def ngram_common(wordlist, ngrams):
    ngram_count = {}
    for i in range(len(wordlist)-ngrams+1):
        ngram = tuple((wordlist[i:i+ngrams]))
        if not ngram in ngram_count:
            ngram_count[ngram] = 1
        else:
            ngram_count[ngram] += 1
    return ngram_count       
        

bigrams = ngram_common(wordlist, ngrams = 2)  
bigrams_sorted = sorted(bigrams.items(), key = lambda x:x[1], reverse = True)
print(f"The most common bigrams are: {bigrams_sorted[0:10]}")
```

```python
# b)
trigrams = ngram_common(wordlist, ngrams = 3)
trigrams_sorted = sorted(trigrams.items(), key = lambda x:x[1], reverse = True)
print(f"\nThe most common trigrams are: {trigrams_sorted[0:10]}")
```

#### 5. Make a Concordance

a. Create a concordance display for an arbitrary word. See the example below  

![](../../Data/figs/Sample-concordance-lines-of-actually.png)

```python
# a)
def concordance(wordlist, random_word, n_of_w_left, n_of_w_right):
    concordance_list = []
    for idx, x in enumerate(wordlist):
        if x in random_word:
            if idx < n_of_w_left: 
                left = wordlist[:idx] 
            else:
                left = wordlist[idx-n_of_w_left:idx] 
            word = x
            right = wordlist[idx+1:idx+n_of_w_right+1]
            concordance_list.append([left,word,right]) 
        
    return(concordance_list)

concordance_display = concordance(wordlist, ["in"], 6, 6)
for x in concordance_display:
    print(x)
```
