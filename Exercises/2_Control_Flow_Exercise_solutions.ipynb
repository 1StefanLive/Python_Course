### 1 - Given the list:

```python
lista = [1,2,3,'4',5,6.3,7.4 + 2j,"123",[1.2,2.5,3.4], 93, "98"]
```

Create a script to print each element of a list, along with its predecessor and its successor. Follow these rules:
+ If this element is an integer, print the value;
+ If it is a string, print the corresponding integer;
+ If it is a real, print the integer part;
+ If it is a complex, print the integer and the imaginary parts;
+ If it is a sequence, apply the rules above for each element of the sequence.

```python
def rules(a):
  if isinstance(a, (str, float, int)):
    b = int(a)
  elif isinstance(a, complex):
    b = int(a.real), a.imag
  return b
```

```python
def transform_to_print(c):
  if isinstance(c, list):
    d = list(map(rules, c))
  else:
    d = rules(c)
  return d
```

```python
for i in range(len(lista)):

    predecessor = transform_to_print(lista[max(0,i-1)])
    element =  transform_to_print(lista[i])
    successor = transform_to_print(lista[min(i+1,len(lista)-1)])

    if i == 0:
      print(element, successor)
    elif i in list(range(1,len(lista)-2)):
      print(predecessor, element, successor)
    elif i == len(lista)-1:
      print(predecessor, element)
```

### Solutions to question number 4

Very slow version

```python
def largest_prime(n):
    #generating primes until n/2
    seq = range(3,int(n/2.)+1)
    primes = [2]
    for candidate in seq:
        is_primo = True
        for prime in primes:
            if candidate%prime == 0:
                is_primo = False
        if is_primo:
            primes.append(candidate)
    factors = []
    for prime in primes:
        if n%prime == 0:
            factors.append(prime)
    print(factors)
    return factors
```

```python
t0 = time.time()
largest_prime(13195)
print(time.time() - t0)
```

```python
def largest_prime_2(n):
    primes = [2]
    prime_candidate = 3
    factors = []
    while prime_candidate < (int(n/2.)+1):
        is_primo = True
        for prime in primes:
            if prime_candidate%prime == 0:
                is_primo = False
        if is_primo:
            primes.append(prime_candidate)
            if n%prime_candidate == 0:
                factors.append(prime_candidate)
        prime_candidate +=1
    print(factors)
    return factors
```

```python
t0 = time.time()
largest_prime_2(13195)
print(time.time() - t0)
```

```python
def is_prime(n):
    if n <= 3:
        return n >= 2
    if n % 2 == 0 or n % 3 == 0:
        return False
    for i in range(5, int(n ** 0.5) + 1, 6):
        if n % i == 0 or n % (i + 2) == 0:
            return False
    return True
```

```python
def largest_prime_3(n):
    factors = []
    prime_candidate = 2
    while prime_candidate < (int(n/2.)+1):
            if is_prime(prime_candidate) and n%prime_candidate == 0:
                factors.append(prime_candidate)
            prime_candidate +=1
    print(factors)
    return factors
```
